```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pickle
```

```python
import pandas as pd

# Load the dataset from the provided CSV file path
file_path = '../Customer_segmentation/OnlineRetail.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset to verify it's loaded correctly
df.head()
```

#### 1. Pre-processing

Pre-processing Steps:

    Load and Clean the Dataset:
        The dataset contains 541,909 rows and 8 columns.
        Columns include InvoiceNo, StockCode, Description, Quantity, InvoiceDate, UnitPrice, CustomerID, and Country.
        There are missing values in the Description and CustomerID columns.

    Handle Missing Values:
        We will drop rows where CustomerID is missing because it is crucial for customer segmentation.
        We will handle missing Description values as they are not critical for the segmentation process.

    Feature Engineering:
        Create a new feature TotalPrice as the product of Quantity and UnitPrice.
        Calculate Recency, Frequency, and Monetary values for each customer:
            Recency: Days since the last purchase.
            Frequency: Total number of purchases.
            Monetary: Total spending.

    Standardization of Features:
        Log-transform the Recency, Frequency, and Monetary features to reduce skewness.
        Standardize these features for clustering.

```python
# Convert InvoiceDate to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Drop rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Remove rows with negative or zero quantities and unit prices
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Create a new feature for TotalPrice
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Calculate Recency, Frequency, and Monetary (RFM) metrics
latest_date = df['InvoiceDate'].max()
rfm_df = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',  # Frequency
    'TotalPrice': 'sum'  # Monetary
}).reset_index()

# Rename columns for clarity
rfm_df.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Log transformation to reduce skewness
rfm_df['Monetary'] = np.log1p(rfm_df['Monetary'])
rfm_df['Recency'] = np.log1p(rfm_df['Recency'])
rfm_df['Frequency'] = np.log1p(rfm_df['Frequency'])

# Standardize the RFM features
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df[['Recency', 'Frequency', 'Monetary']])

# Display the first few rows of the processed data
rfm_df.head()
```

The pre-processing steps have been successfully completed, and we now have a clean and processed dataset with the following features:

    Recency: Log-transformed days since the last purchase.
    Frequency: Log-transformed number of unique transactions.
    Monetary: Log-transformed total spending by the customer.

These features have been standardized

#### 3. Clustering with K-Means

```python
# Determine the Optimal Number of Clusters
# Elbow Method to find the optimal number of clusters.
```

```python
from sklearn.cluster import KMeans

# Prepare the RFM features from the previous steps (already calculated)
# Standardize the RFM features
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df[['Recency', 'Frequency', 'Monetary']])

# Elbow method to find the optimal number of clusters
sse = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
#The "elbow" point in the plot suggests the optimal number of clusters to use for K-Means clustering. Typically, the point where the SSE starts to decrease at a slower rate indicates the best balance between model complexity and accuracy.
#Interpretation:

   # Elbow Point: In the plot, the "elbow" is typically identified where the reduction in SSE begins to level off. Based on the plot, it looks like the optimal number of clusters might be around 4 or 5 clusters, where the curve starts to bend and the decrease in SSE becomes less significant.

#Next Steps:

   # Choosing the Number of Clusters: Based on plot, choose 4 or 5 clusters for your K-Means model.
    #Train the K-Means Model:  proceed to train the K-Means model using that number of clusters.
```

#### train the k-means model

```python
# Let's choose 4 as the optimal number of clusters based on the Elbow plot
optimal_k = 4

# Train the K-Means model
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Display the first few rows with cluster assignments
rfm_df.head()
```

#### simulation function to make prediction

```python
# Function to predict the cluster of a customer by CustomerID
def predict_customer_segment_by_id(customer_id):
    customer_data = rfm_df[rfm_df['CustomerID'] == customer_id]
    if customer_data.empty:
        print(f"No customer found with CustomerID {customer_id}")
        return None
    
    scaled_data = scaler.transform(customer_data[['Recency', 'Frequency', 'Monetary']])
    cluster_label = kmeans.predict(scaled_data)
    return cluster_label[0]

# Example: Predict cluster for a specific CustomerID
customer_id = 12347.0  # Replace this with any valid CustomerID from the dataset
predicted_cluster = predict_customer_segment_by_id(customer_id)
print(f'Customer with ID {customer_id} belongs to cluster {predicted_cluster}')
```

```python
# Example CLTV calculation for a given cluster
cluster_id = 2  # Choose a cluster
avg_monetary = rfm_df[rfm_df['Cluster'] == cluster_id]['Monetary'].mean()
avg_frequency = rfm_df[rfm_df['Cluster'] == cluster_id]['Frequency'].mean()
avg_recency = rfm_df[rfm_df['Cluster'] == cluster_id]['Recency'].mean()

# CLTV formula using average values from the cluster
cltv = avg_monetary * avg_frequency / avg_recency
print(f'Estimated CLTV for cluster {cluster_id}: {cltv:.2f}')
```

```python
# Define churn risk based on high recency and low frequency/monetary values
rfm_df['ChurnRisk'] = rfm_df.apply(
    lambda x: 'High' if (x['Recency'] > np.log1p(30)) and (x['Frequency'] < np.log1p(2)) else 'Low', axis=1
)

# Count churn risk by cluster
churn_risk_counts = rfm_df.groupby('Cluster')['ChurnRisk'].value_counts().unstack().fillna(0)
print(f'Churn Risk by Cluster:\n{churn_risk_counts}')
```
